```python
import pybedtools
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys

#Redo Figure 5
```

```python
def makeFig5(TEtool):
    #Merge the df from Fig5 (contains the motifs) with the TE overlap info from the four tools and replot!
    plt.clf()
    df = pd.read_csv("Filtered_de_novo_transcripts_homologs.csv")

    df[["ID", "Other"]] =  df["Transcript_Id"].str.split("::", n=1, expand=True)
    df.loc[df["Transcription_status"]=="Transcript","ID"] = df["ID"] + "_" + df["Other"] + "_Transcript"
    df.loc[df["Transcription_status"]=="Non_transcribed_Homolog","ID"] = "NonExpressedHomolog_" + df["ID"] + "::" + df["Other"] + "_"+ df["Population"] + "_" +  df["Other"] + "_"+ df["Population"] + "_Transcript"

    te = pd.read_csv(f"Figure3/Fig1C_Dataframe_{TEtool}.csv")
    x = pd.merge(df, te, on="ID", how = "left")

    x["TypeTE"] = "TEs overlap"
    x.loc[x["Type"]=="NoTE","TypeTE"] = "No TE overlap"

    ax = sns.pointplot(data=x, x = "TypeTE", y = "Number_motifs", hue ="Transcription_status", errorbar=("ci", 100), capsize=.4, linestyle="none", marker="D", dodge = 0.3, palette = ["#953553", "lightgreen"], order = ["No TE overlap", "TEs overlap"])
    ax.set(title = f"TEtool = {TEtool}")
    plt.savefig(f"{TEtool}_Fig5.jpg")
    x.to_csv(f"{TEtool}Data_Fig5.csv") #Save the df
```
