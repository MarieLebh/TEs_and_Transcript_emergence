```python
import pybedtools
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
import numpy as np

#Redo Figure 4
```

```python
def RedoFig4C_D(TEtool):
    #Redo the figure with the motifs 
    df = pd.read_csv("Data_motifs_new.csv")

    df[["ID", "Other"]] =  df["unique_id"].str.split("::", n=1, expand=True)
    df.loc[df["type"]=="transcript","type"] = "Transcript"
    df["ID"] = df["ID"]+"_" + df["population"] +"_" + df["type"]

    tr = pd.read_csv(f"Fig1A_Dataframe_{TEtool}_allseqs.csv")
    x = pd.merge(df, tr, on="ID", how = "left")

    x.loc[x["type"]!="Transcript","TEpresence"] =x["type"]
    x.to_csv(f"{TEtool}_dataFig4_motif.csv")
    fig, axs = plt.subplots(ncols=2, figsize=(11,4))
    ax = sns.pointplot(
        data=x, x="TEpresence", y="number_motifs_per_transcript", hue = "TEpresence",legend = None, 
        errorbar=("ci", 100), capsize=.4, color=".5", linestyle="none", marker="D",order = ["dn_no_TE", "dn_overlap_TE", "dn_TE", "gene", "Intergenic"], hue_order = ["dn_no_TE", "dn_overlap_TE", "dn_TE", "gene", "Intergenic"],ax = axs[0],palette = ["darkgreen", "green", "lightgreen","orange","blue"])
    ax.set(title = "Low identity score")
    ax.set_ylabel("# Motif")
    ax.set_xlabel(" ")

    ax = sns.pointplot(
        data=x, x="TEpresence", y="number_motifs_high_per_transcript",hue = "TEpresence", legend = None, 
        errorbar=("ci", 100), capsize=.4,
        color=".5", linestyle="none", marker="D", order = ["dn_no_TE", "dn_overlap_TE", "dn_TE", "gene", "Intergenic"],hue_order = ["dn_no_TE", "dn_overlap_TE", "dn_TE", "gene", "Intergenic"],ax = axs[1],palette = ["darkgreen", "green", "lightgreen","orange","blue"])
    ax.set(title = "High identity score")
    ax.set_ylabel("# Motif")
    ax.set_xlabel(" ")
    plt.tight_layout()

    plt.suptitle(f"TEtool = {TEtool}", fontsize=12)
    plt.savefig(f"{TEtool}4AB.jpg")
```
